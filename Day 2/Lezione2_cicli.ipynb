```python
counter= 0
while counter < 5:
    print(counter)
    counter +=1
    if (counter ==3):
        break
        print (counter)
```

```python
counter= 0
while counter < 5:
    print(f"within the loop:  {counter}") #cpunter inside the loop
    counter +=1
    if (counter ==3):
        break
        
print (f"outside of the loop: {counter}") #counter outside the loop
```
